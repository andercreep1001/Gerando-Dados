```python
import matplotlib.pyplot as plt

#valor_x = [1,2,3,4,5]
valor_x = range(1, 1001)
valor_y = [x**2 for x in valor_x]
#valor_y = [1,4,9,16,25]

plt.style.use('seaborn') #cria um plano de fundo

fig, ax = plt.subplots()

ax.scatter(valor_x, valor_y, c=valor_y, cmap=plt.cm.Blues, s=100)

ax.set_title("Numeros quadrados", fontsize=24)
ax.set_xlabel("Valor", fontsize=15)
ax.set_ylabel("Valor quadrado", fontsize=15)

ax.tick_params(axis='both', which='major', labelsize=15)

ax.axis([0, 1100, 0, 1100000])

plt.show()
```
